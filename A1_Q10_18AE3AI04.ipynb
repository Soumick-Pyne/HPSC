```python
import pandas as pd
import wikipedia

articles = ['Linear algebra', 'Data Science', 'Artificial intelligence', 'European Central Bank',
            'Financial technology', 'International Monetary Fund', 'Basketball', 'Swimming',
            'Cricket']
wiki_list = []
title = []

for article in articles:
  print("loading content : ", article)
  wiki_list.append(wikipedia.page(article).content)
  title.append(article)
print("Examine Content")
wiki_list
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words={'english'})
X = vectorizer.fit_transform(wiki_list)
```

```python
print(X.shape)
```

```python
print(X)
```

**Kmeans.inertia_** is the Sum of squared distances of samples to their closest cluster center

```python
from sklearn.cluster import KMeans as kmeans
loss_arr = []
k_values = [4, 6, 8]

for k in k_values:
  k_obj = kmeans(n_clusters = k, max_iter = 140)
  k_obj = k_obj.fit(X)
  loss_arr.append(k_obj.inertia_)
  print("For K =", k, "the loss (sum of square distances) = ", round(k_obj.inertia_, 4), "\n")
  #K=12 will not make sense since we have 9 datapoints (N = 9).
```

```python
for k in k_values:
  clusters = kmeans(n_clusters = k, max_iter = 100)
  clusters.fit(X)
  labels = clusters.labels_
  wiki = pd.DataFrame(list(zip(title, labels)), columns=['title', 'cluster'])
  print("Document-Cluster association for K = ", k, "\n", wiki, "\n\n")
```

For k=4 the algorithm successfully generalises to group together the sports document under one class, but it also includes documents on financial intitutions and AI in the same category. <br/>

For k=6 our objection with k=4 case still persists. Sports and Financial Institution related documents are in same cluster. Though it managed to create a new cluster for AI. k=6 would have been acceptable if Finacial Intitution documents and sports documents are in seperate clusters, and instead ML and data science documents were grouped topgether. <br/>

K=8, though not ideal (bad generalisation), is still the best out of the three in terms of keeping documents of completely different topics in seperate clusters.
